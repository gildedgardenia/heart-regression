```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sb

heart_df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
print(heart_df.info())

# split into survivors and deaths

survivors_df = heart_df[heart_df['DEATH_EVENT'] == 0]
deaths_df = heart_df[heart_df['DEATH_EVENT'] == 1]
```

```python
survivors_df = survivors_df.drop(['diabetes', 'sex', 'smoking', 'time', 'DEATH_EVENT'], axis=1)
```

```python
sb.heatmap(survivors_df.corr(), cmap="YlGnBu", annot = True)
plt.show()
```

```python
sb.pairplot(survivors_df, x_vars=['serum_creatinine', 'ejection_fraction', 'anaemia'], 
             y_vars='age', size=4, aspect=1, kind='scatter')
plt.show()
```

```python
deaths_df = deaths_df.drop(['high_blood_pressure', 'diabetes', 'sex', 'smoking', 'time', 'DEATH_EVENT'], axis=1)
```

```python
sb.heatmap(deaths_df.corr(), cmap="YlGnBu", annot = True)
plt.show()
```

```python
sb.pairplot(survivors_df, x_vars=['serum_creatinine', 'age', 'serum_sodium'], 
             y_vars='ejection_fraction', size=4, aspect=1, kind='scatter')
plt.show()
```
